```python
import numpy as np
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from pylab import *

 
x=-4.9999
dx=0.0001
f0, f1 = 0, 0.00001

p=np.zeros(100)
phi=np.zeros(100)
p[0]=-5
def g(E)
    E=6.5
    return (E)
for m in range (2,100000,1):
    x=x+dx
    f2=2*f1-f0+f1*(x**2-2*E)*dx**2
    if m %1000==0:
        xi=int(m/1000)
        p[xi]=x
        phi[xi]=f2
    f0=f1
    f1=f2
plot(p,phi)
show()

```
